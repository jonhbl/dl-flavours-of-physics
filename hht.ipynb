```python
import librosa
import matplotlib.pyplot as plt
import numpy as np
import scipy.signal as signal
import emd
```

### Read File


```python
data, fs = librosa.load("audio.mp3")
```

```python
len(data) / fs  # duration in seconds: 1min 46s
```

```python
plt.figure(figsize=(15, 5))
plt.plot(data)
plt.title("Audio Signal")
plt.xlabel("Samples")
plt.ylabel("Amplitude")
plt.show()
```

### Cut file


```python
size_in_seconds = 10
size_in_samples = size_in_seconds * fs
```

```python
data = data[:size_in_samples]
```

### EMD


```python
data_emd = emd.sift.sift(data)
```

```python
emd.plotting.plot_imfs(data_emd)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(len(data_emd.T), 1, figsize=(15, 20))
for i, coef in enumerate(data_emd.T):
    ax[i].plot(coef)
    ax[i].set_title(f"IMF {i+1}")
    ax[i].grid()

plt.tight_layout()
```

A HHT extrai o conjunto de componentes oscilatórios de um sinal, representados pelos IMFs.
<br>
Ainda que a interpretabilidade da técnica seja similar à DWT, com os primeiros componentes de mais oscilação e, consequentemente, frequência, nota-se que a HHT é capaz de extrair uma tendência mais clara nas frequências mais baixas.
<br>
A DWT apresenta, até na sua última camada, oscilações mais rápidas, por analisar diretamente a banda de frequência que sobrou.
<br>
A HHT, por sua vez, dada a abordagem empírica e adaptativa, apresenta uma decomposição mais representativa nas baixas frequências da tendência do sinal.
